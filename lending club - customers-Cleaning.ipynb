## 1.Creating Data Types

```
customers_schema='member_id string, emp_title string, emp_length string, home_ownership string, annual_inc float, addr_state string, zip_code string, country string, grade string, sub_grade string, verification_status string,tot_hi_cred_lim float, application_type string, annual_inc_joint float, verification_status_joint string'
```

```
customer_raw_df=spark.read\
    .format("csv")\
    .option("header","true")\
    .schema(customers_schema)\
    .load("dbfs:/user/lendingclub_rawfiles/temp/customer_data_csv")
```

```
customer_raw_df.printSchema()
```

## 2.Renaming Columns

```
customer_df_renamed=customer_raw_df.withColumnRenamed("annual_inc","annual_income")\
    .withColumnRenamed("addr_state","address_state")\
    .withColumnRenamed("zip_code","address_zipcode")\
    .withColumnRenamed("country","address_country")\
    .withColumnRenamed("tot_hi_cred_lim","total_high_credit_limit")\
    .withColumnRenamed("annual_inc_joint","joint_annual_income")
```

```
customer_df_renamed.display()
```

## 3.Ingesting Current TimeStamp

```
from pyspark.sql.functions import *
customer_df_ingested= customer_df_renamed.withColumn("ingest_date",current_timestamp())
```

```
customer_df_ingested.display()
```

```
customer_df_ingested.printSchema()
```

```
customer_df_ingested.count()
```

## 4.Removing Duplicates

```
customer_distinct_df= customer_df_ingested.distinct()
```

```
customer_distinct_df.count()
```

```
customer_df_ingested.createOrReplaceTempView("customers")
```

## 5. Removing Rows where annual_inc is null

```
customer_income_filtered=spark.sql("select * from customers where annual_income is not null")
```

```
customer_income_filtered.createOrReplaceTempView("customers")
```

```
%sql
select distinct(emp_length) from customers
```

```
from pyspark.sql.functions import regexp_replace,col
customer_emplength_cleaned= customer_income_filtered.withColumn("emp_length", regexp_replace(col("emp_length"),"(\D)",""))
```

```
customer_emplength_cleaned.display()
```

##6.coverting emp_length to integer

```
customer_emplength_casted=customer_emplength_cleaned.withColumn("emp_length",customer_emplength_cleaned.emp_length.cast('int'))
```

```
customer_emplength_casted.printSchema()
```

```
customer_emplength_casted.filter("emp_length is null").count()
```

```
customer_emplength_casted.createOrReplaceTempView("customers")
```

##7. Replacing Emp_Length Nulls with avg value

```
avg_emp_length=spark.sql("select floor(avg(emp_length)) from customers").collect()
```

```
avg_emp_duration=avg_emp_length[0][0]
print(avg_emp_duration)
```

```
customer_emplength_replaced=customer_emplength_casted.na.fill(avg_emp_duration,subset=['emp_length'])
```

```
customer_emplength_replaced.display()
```

```
customer_emplength_replaced.filter("emp_length is null").count()
```

```
customer_emplength_replaced.createOrReplaceTempView("customers")
```

```
%sql
select distinct(address_state) from customers
```

```
%sql
select count(address_state) from customers where length(address_state)>2
```

## 8. Cleaning Address_State and filling nulls inplace of more than 2 chars

```
    from pyspark.sql.functions import when, col, length
    customer_state_cleaned=customer_emplength_replaced.withColumn("address_state",
                                           when(length(col("address_state"))>2,"NA").otherwise(col("address_state")))
```

```
customer_state_cleaned.select("address_state").distinct()
```

```
customer_state_cleaned.createOrReplaceTempView("customers")
```

```
%sql
select address_state from customers where address_state is null
```

##9. Writing and saving to disk

```
customer_state_cleaned.write\
    .format("parquet")\
    .mode("overwrite")\
    .option("path","dbfs:/user/lendingclubproject/cleaned/customers_parquet")\
    .save()
```

```
customer_state_cleaned.write\
    .format("csv")\
    .mode("overwrite")\
    .option("path","dbfs:/user/lendingclubproject/cleaned/customers_csv")\
    .save()
```
