#1.Importing the DATA

```
raw_df=spark.read\
    .format("csv")\
    .option("header","true")\
    .option("inferSchema","true")\
    .load("dbfs:/FileStore/accepted_2007_to_2018Q4.csv")
```

#2.Creating the temp table on top of DF

```
raw_df.createOrReplaceTempView("lendingnewtable")
```

#4.Assiging the Uniq id using some coloumns

```
from pyspark.sql.functions import sha2,concat_ws
```

```
new_df=raw_df.withColumn("name_sha2",sha2(concat_ws("||",*["emp_title","emp_length","home_ownership","annual_inc","zip_code","addr_state","grade","sub_grade","verification_status"]),256))
```

```
new_df.createOrReplaceTempView("lendingnewtable")
```

#Writing data to customer 

```
# SQL query to select data from your table
result_df = spark.sql("""
    SELECT 
        name_sha2 AS member_id, 
        emp_title, 
        emp_length, 
        home_ownership, 
        annual_inc, 
        addr_state, 
        zip_code, 
        'USA' AS country, 
        grade, 
        sub_grade, 
        verification_status,
        tot_hi_cred_lim, 
        application_type, 
        annual_inc_joint, 
        verification_status_joint 
    FROM lendingnewtable
""")

# Limit the number of rows to, for example, 100
result_df = result_df.limit(100)

# Saving the limited DataFrame as a CSV file
result_df.repartition(1).write\
    .option("header", "true")\
    .format("csv")\
    .mode("overwrite")\
    .option("path", "dbfs:/user/lendingclub_rawfiles/temp/customer_data_csv")\
    .save()
```

```
result_df.display()
```

#writing data to loans df

```
# SQL query to select data from your table
loans_df = spark.sql("""
    SELECT 
        id AS loan_id, 
        name_sha2 AS member_id, 
        loan_amnt, 
        funded_amnt, 
        term, 
        int_rate, 
        installment, 
        issue_d, 
        loan_status, 
        purpose, 
        title 
    FROM lendingnewtable
""")

# Limit the number of rows to, for example, 100 (you can change the number)
loans_df = loans_df.limit(100)

# Saving the limited DataFrame as a CSV file
loans_df.repartition(1).write\
    .option("header", "true")\
    .format("csv")\
    .mode("overwrite")\
    .option("path", "dbfs:/user/lendingclub_rawfiles/temp/loans_data_csv")\
    .save()
```

```
loans_df.display()
```
